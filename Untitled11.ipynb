```python
import dash
from dash import html,dcc

from dash.dependencies import Input, Output
import plotly.graph_objects as go

# Initialize the Dash app
app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("Waterfall Chart"),
    dcc.Graph(id='waterfall-chart'),
])

@app.callback(
    Output('waterfall-chart', 'figure'),
    [Input('waterfall-chart', 'relayoutData')]
)
def update_graph(relayoutData):
    # Define the Waterfall chart data
    measure = ["absolute", "relative", "absolute", "relative", "absolute"]
    x_data = ["Start", "Relative 1", "Mid", "Relative 2", "End"]
    y_data = [10, 5, 20, -5, 15]  # Example data

    # Create the waterfall chart
    figure = go.Figure(go.Waterfall(
        name = "Waterfall Chart",
        orientation = "v",
        x = x_data,
        y = y_data,
        measure = measure,
        textposition = "outside",
        text = [f"{y}" for y in y_data]
    ))

    # Calculate the difference between bars
    diff1 = y_data[2] - y_data[0]
    diff2 = y_data[4] - y_data[2]

    # Midpoint calculation
    midpoint1 = (x_data.index("Start") + x_data.index("Mid")) / 2
    midpoint2 = (x_data.index("Mid") + x_data.index("End")) / 2

    # Add circles to display the difference
    figure.add_trace(go.Scatter(
        x=[x_data[int(midpoint1)], x_data[int(midpoint2)]],
        y=[max(y_data) + 5, max(y_data) + 5],  # Place the circle above the bars
        text=[f"Diff: {diff1}", f"Diff: {diff2}"],
        mode='text',
        textfont=dict(size=12, color="black")
    ))

    # Update the layout
    figure.update_layout(title="Waterfall Chart with Differences Displayed")

    return figure

if __name__ == '__main__':
    app.run_server()
```
